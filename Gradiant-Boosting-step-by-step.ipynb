```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
df['X'] =  X.reshape(100)
df['y'] =  y
df
```

```python
plt.scatter(df['X'], df['y'])
```

```python
df['pred-1'] = df["y"].mean()
df
```

```python
df['res-1'] = df['y'] - df['pred-1']
```

```python
plt.scatter(df['X'], df['y'])
plt.plot(df['X'], df['pred-1'], color='red')
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree1 = DecisionTreeRegressor(max_leaf_nodes=8)
tree1.fit(df[['X']].values.reshape(100, 1), df['res-1'].values)
```

```python
from sklearn.tree import plot_tree

plot_tree(tree1)
```

```python
X_test = np.linspace(-0.5, 0.5, 500)
```

```python
y_pred = df['pred-1'][1] + tree1.predict(X_test.reshape(500, 1))
```

```python
plt.figure(figsize=(14, 4))
plt.subplot(121)
plt.plot(X_test, y_pred, linewidth=2, color='red')
plt.scatter(df['X'], df['y'])
```

```python
df['pred-2'] = df['pred-1'] + tree1.predict(df['X'].values.reshape(100, 1))
df
```

```python
df['res-2'] = df['y'] - df['pred-2']
df
```

```python
tree2 = DecisionTreeRegressor(max_leaf_nodes=8)
tree2.fit(df['X'].values.reshape(100, 1), df['res-2'].values)
```

```python
plot_tree(tree2)
```

```python
y_pred = df['pred-1'][1] + sum(regressor.predict(X_test.reshape(-1,1)) for regressor in [tree1, tree2])
plt.figure(figsize=(14, 4))
plt.subplot(121)
plt.plot(X_test, y_pred, linewidth=2, color='red')
plt.scatter(df['X'], df['y'])
```
