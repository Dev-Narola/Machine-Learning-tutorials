```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
from sklearn.datasets import load_digits

data = load_digits()
data
```

```python
np.unique(data.target)
```

```python
columns_name = list(data.feature_names)
columns_name.append('target')
df = pd.DataFrame(np.c_[data.data, data.target], columns=columns_name)
df
```

```python
x = df.iloc[:,0:-1]
y = df.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scale = scaler.fit_transform(x)
x_scale
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scale, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
lr.score(x_test, y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
x_train_pca.shape
```

```python
lr = LogisticRegression()
lr.fit(x_train_pca, y_train)
lr.score(x_test_pca, y_test)
```

```python
y_pred = lr.predict(x_test_pca)
```

```python
accuracy_score(y_test, y_pred)
```

```python
for i in range (1, 65):
    pca = PCA(n_components=i)
    x_train_pca = pca.fit_transform(x_train)
    x_test_pca = pca.transform(x_test)
    lr = LogisticRegression()
    lr.fit(x_train_pca, y_train)
    lr.score(x_test_pca, y_test)
    y_pred = lr.predict(x_test_pca)
    print(accuracy_score(y_test, y_pred))
```

```python
pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
```
